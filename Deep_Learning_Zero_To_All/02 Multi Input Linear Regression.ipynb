### 01. How to define numpy,matplotlib and tensorflow?

```python
import numpy as np
from matplotlib import pyplot as plt
import tensorflow as tf
```

### 02. How to make Multi Input train value?

```python
x_train = [[1,2,3],
           [2,3,4],
           [3,4,5],
           [4,5,6]]
y_train = [[1.1],
           [2.1],
           [3.1],
           [4.1]]
```

### 03. How to define model?

```python
model = tf.keras.models.Sequential()
```

```python
model.add(tf.keras.layers.Dense(units=1,input_dim=3))
```

### 04. How to define schastic gradient descent?

```python
sgd = tf.keras.optimizers.SGD(lr=0.001)
```

### 05. How to compile ?

```python
model.compile(optimizer=sgd,loss='mse')
```

#### 06. How to visualize summary?

```python
model.summary()
```

### 07. How to do fitting?

```python
history = model.fit(x_train,y_train,epochs=1000)
```

```python
model.predict([[1,2,3]])
```

```python
plt.plot(history.history['loss'])
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.grid()
```
