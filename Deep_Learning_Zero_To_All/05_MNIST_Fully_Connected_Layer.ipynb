### 01. How to define numpy,matplotlib and tensorflow?

```python
import numpy as np
from matplotlib import pyplot as plt
import tensorflow as tf
```

### 02. How to load MNIST dataset?

```python
mnist = tf.keras.datasets.mnist
```

```python
(x_train,y_train),(x_test,y_test) = mnist.load_data()
```

### 03. How to normalize?

```python
np.float(np.max(x_train))
```

```python
x_train = x_train/np.float(np.max(x_train))
x_test = x_test/np.float(np.max(x_test))
```

### 04. How to increase train dataset axis?

```python
x_train.shape
```

```python
x_train = x_train.reshape(x_train.shape[0],x_train.shape[1]*x_train.shape[2])
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
x_test = x_test.reshape(x_test.shape[0],x_test.shape[1]*x_test.shape[2])
```

```python
y_test.shape
```

### 05. How to define model?

```python
model = tf.keras.models.Sequential()
```

```python
model.add(tf.keras.layers.Dense(units=10,input_dim=784))
```

```python
model.add(tf.keras.layers.Activation('softmax'))
```

### 06. How to define adam for optimizer?

```python
adam = tf.keras.optimizers.Adam(0.1)
```

### 7. How to compile(not one-hot encoding) ?

```python
model.compile(optimizer=adam,loss='sparse_categorical_crossentropy',metrics=['acc'])
```

### 06. How to visualize summary?

```python
model.summary()
```

### 07. How to do fitting?

```python
history = model.fit(x_train,y_train,epochs=3)
```

```python
model.predict(x_test)[0]
```

### 08. How to draw loss graph?

```python
plt.plot(history.history['loss'])
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.grid()
```

### 09. How to draw accuracy graph?

```python
plt.plot(history.history['acc'])
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.grid()
```

### 10.How to evaluate ?

```python
model.evaluate(x_test,y_test)
```
