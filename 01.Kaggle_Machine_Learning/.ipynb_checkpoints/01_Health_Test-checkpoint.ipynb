## 01.01 데이터 결합

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
import statsmodels.formula.api as smf
from statsmodels.formula.api import ols, glm
```

```python
import glob
import os

all_files = glob.glob(os.path.join('HEALTH*'))
all_files
```

```python
all_data_frames = []
for file in all_files:
    data_frame = pd.read_csv(file, index_col=None, sep=',', header=0)
    all_data_frames.append(data_frame)
all_data_frames
```

```python
health = pd.concat(all_data_frames, axis=0, ignore_index=True) #axis=0 수직, axis=1 수평

health

health.to_csv("health_both.csv", index = False)
```

```python
health = health.dropna()
```

### - 200 sample만 선택하여 상관관계 그래프 그리기

```python
def sampling(data, n = 200):
    return data.loc[np.random.choice(data.index,size=n)]
health_sample = sampling(health)
```

### -sex(성) 는 correlation value는 높게 나오지만, 그래프를 보았을 때 상관성이 있다고 판단하기 어려움 

```python
sns.pairplot(health_sample,vars=['bp_high','bp_lwst','waist','age_group','weight','sex'],kind='reg')#,hue='type')
```

## - 평균적으로 1(남자) 가 2(여자) 보다 혈압이 높으며 분산도 작음 
## - 대체적으로 남자가 고혈압이라고 할 수 있으므로 포함

# 02. 학습 및 평가를 위한 데이터셋을 Train/Validation/Test 

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
x_value = ['bp_lwst','waist','age_group','sex','weight'] 
y_value = ["bp_high"]

x_datas = health[x_value]
y_datas = health[y_value]

x_train, x_test, y_train, y_test = train_test_split(x_datas, y_datas,random_state=42,test_size=0.1) #8:1:1
x_train, x_valid, y_train, y_valid = train_test_split(x_train,y_train,test_size=1/9) #
```

```python
print("x_train.shape =",x_train.shape," y_train.shape =",y_train.shape)
```

```python
print("x_valid.shape =",x_valid.shape," y_valid.shape =",y_valid.shape)
```

```python
print("x_test.shape =",x_test.shape," y_test.shape =",y_test.shape)
```

# 3. 평가

## 4개의 모델(기본 hyper parameter) 해본 결과 Randomforest가 가장 좋은 모델임을 확인

```python
from sklearn.linear_model import LinearRegression 
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn import metrics  


model1 = LinearRegression()
model1.fit(x_train,y_train)
prediction1 = model1.predict(x_valid)
score1 = np.sqrt(metrics.mean_squared_error(y_valid, prediction1))

model2 = RandomForestRegressor()
model2.fit(x_train,y_train)
prediction2 = model2.predict(x_valid)
score2 = np.sqrt(metrics.mean_squared_error(y_valid, prediction2))

model3 = DecisionTreeRegressor()
model3.fit(x_train,y_train)
prediction3 = model3.predict(x_valid)
score3 = np.sqrt(metrics.mean_squared_error(y_valid, prediction3))

model4 = GradientBoostingRegressor()
model4.fit(x_train,y_train)
prediction4 = model4.predict(x_valid)
score4 = np.sqrt(metrics.mean_squared_error(y_valid, prediction4))

print("LinearRegression=",score1,"RandomForestRegressor=",score2,"DecisionTreeRegresso=",score3,"GradientBoostingRegressor",score4)
```

## 4. 최적화

## 3번에서 선택한 randomforest 모델활용, 
## grid search 방법으로 hyper parameter 선정하여 최적화

```python
from sklearn.linear_model import LinearRegression ## 회귀알고리즘 추가
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics  

n_estimators = 100

max_depth_list = [15,18,21]
max_features_list = [0.3,0.5,0.7]

hyperparameters_list=[]
for max_depth in max_depth_list :
    for max_features in max_features_list :
        model = RandomForestRegressor(n_estimators= n_estimators,
                                     max_depth = max_depth,
                                     max_features= max_features,
                                     random_state=10,
                                     n_jobs=-1)
        model.fit(x_train,y_train)
        prediction = model.predict(x_valid)
        score = np.sqrt(metrics.mean_squared_error(y_valid, prediction))
        
        hyperparameters_list.append({
            'score':score,
            'max_depth':max_depth,
            'max_feature':max_features
        })
        
        print("score = {}".format(score))
        
```

## - 점수가 가장 낮은  {'score': 7.431861243396372, 'max_depth': 21, 'max_feature': 0.3}, 가 최적

```python
hyperparameters_list
```

```python
model = RandomForestRegressor(n_estimators= n_estimators,
                                     max_depth = 21,
                                     max_features= 0.3,
                                     random_state=10,
                                     n_jobs=-1)
model.fit(x_train,y_train)
prediction = model.predict(x_test)
score = np.sqrt(metrics.mean_squared_error(y_test, prediction))
        

print(score)
```
