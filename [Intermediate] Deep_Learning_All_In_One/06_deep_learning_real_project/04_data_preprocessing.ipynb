### 04.02 Define Class name

```python
class_list = set()
for new_image in new_images:
    name = new_image.split('/')
    name = name[-1].split('.')[-2]
    name = name.split('_')[0]
    class_list.add(name)
class_list = list(class_list)
class_list = class_list
```
