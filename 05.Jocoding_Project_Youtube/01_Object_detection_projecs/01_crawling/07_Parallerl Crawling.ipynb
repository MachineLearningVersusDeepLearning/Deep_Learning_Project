```python
#!/usr/bin/python3
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
import urllib.request
import os
```

# Environment Setting

```python
binary_location = "/usr/bin/google-chrome"
driver_location = "/usr/local/bin/chromedriver"
```

# Calculate the number of folders

```python
real_path = os.path.realpath("/home/nvidia/ai_data/")
path = real_path + "/entertainer_picture"
number = 0

####################################### find the character folder ##########################
for folder in os.listdir(path):
    ################################## make full path ###########################
    character_path = path+"/"+str(folder)
    ##################################  find the sex folders #####################
    sexs = os.listdir(character_path)
    for sex in sexs:
        sex_path = character_path +"/" + str(sex)
        ############################ find the persone name foldders #############3
        names = os.listdir(sex_path)
        for name in names:
            number += 1
            _string = "google_{}".format(number)
            print(_string)
            globals()['google_%s' %number] =   webdriver.Chrome(executable_path = driver_location,chrome_options=option)
            globals()['google_%s' %number].get("http://www.google.co.kr")

            
            ############################## find the search menu ################################
            elem = globals()['google_%s' %number].find_element_by_name("q")
            ############################## find the word ##################################
            elem.send_keys(name)
            ################################ enter ################################
            elem.send_keys(Keys.RETURN)
            ################################  click the image ################################
            globals()['google_%s' %number].find_element_by_class_name('hide-focus-ring').click()
            time.sleep(1)


            ############################ pause time  ################################
            SCROLL_PAUSE_TIME = 3

            ########################### scroll down code ########################
            # Get scroll height
            last_height = globals()['google_%s' %number].execute_script("return document.body.scrollHeight")
        
            while True:
                # Scroll down to bottom
                globals()['google_%s' %number].execute_script("window.scrollTo(0, document.body.scrollHeight);")
            
                ############################ Wait to load page#############
                time.sleep(SCROLL_PAUSE_TIME)
            
                # Calculate new scroll height and compare with last scroll height
                new_height = globals()['google_%s' %number].execute_script("return document.body.scrollHeight")
                if new_height == last_height:
                    try:
                        globals()['google_%s' %number].find_element_by_class_selector('.mye4qd').click
                    except:
                        print("Done")
                        break
                last_height = new_height

                ############################## click the picture ##########
                images = globals()['google_%s' %number].find_elements_by_css_selector(".rg_i.Q4LuWd")
                for i,image in enumerate(images):
                    try:
                        image.click()
                        time.sleep(2)
                        ############################### src #############################
                        url = globals()['google_%s' %number].find_element_by_xpath('/html/body/div[2]/c-wiz/div[3]/div[2]/div[3]/div/div/div[3]/div[2]/c-wiz/div[1]/div[1]/div/div[2]/a/img').get_attribute("src") #  css_selector(".n3VNCb").get_attribute("src")
                        ############################ download #########################
                        r = urllib.request.urlretrieve(url,name_path+"/"+str(i)+".jpg")
                    except:
                        print("error")
                        
            globals()['google_%s' %number].close()
```
