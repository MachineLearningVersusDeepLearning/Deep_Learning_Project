## 웹 자동화에 대부분 사용하고 있음

# 01. 라이브러리 불러오기

```python
from selenium import webdriver
```

```python
browser = webdriver.Chrome()
```

```python
browser.get("http://naver.com")
```

# 02. 로그인창 클릭하기

```python
elem = browser.find_element_by_class_name('link_login')
elem.click()
```

# 03. id_passwd 입력하기

```python
elem = browser.find_element_by_id("id")
elem.send_keys("id")
```

```python
elem = browser.find_element_by_id("pw")
elem.send_keys("pd")
```

```python
elem = browser.find_element_by_class_name('btn_global')
elem.click()
```

# 04. html 파일 읽기

```python
browser.page_source
```

# 05. 종료

```python
browser.quit()
```
